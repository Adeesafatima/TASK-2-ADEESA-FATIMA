```python
from datetime import datetime
 
LINE  = "=" * 50
DLINE = "-" * 50
 
def print_header():
    print(LINE)
    print("       DECODELABS — EXPENSE TRACKER v2.0")
    print("         Python Programming | Project 2")
    print(LINE)
    print(f"  Session started : {datetime.now().strftime('%d %b %Y  %I:%M %p')}")
    print(LINE)
    print("  Commands:")
    print("    Enter a number  → add expense")
    print("    'history'       → view all entries")
    print("    'undo'          → remove last entry")
    print("    'reset'         → clear everything")
    print("    'quit'          → finish & show report")
    print(LINE)
    print()
 
def print_report(expenses):
    print()
    print(LINE)
    print("            FINAL EXPENSE REPORT")
    print(LINE)
 
    if not expenses:
        print("  No expenses were recorded.")
    else:
        total   = sum(expenses)
        count   = len(expenses)
        avg     = total / count
        largest = max(expenses)
        smallest= min(expenses)
 
        print(f"  Total entries    : {count}")
        print(f"  Largest expense  : Rs. {largest:,.0f}")
        print(f"  Smallest expense : Rs. {smallest:,.0f}")
        print(f"  Average expense  : Rs. {avg:,.2f}")
        print(DLINE)
        print(f"  TOTAL SPENT      : Rs. {total:,.2f}")
 
    print(LINE)
    print(f"  Session ended : {datetime.now().strftime('%d %b %Y  %I:%M %p')}")
    print(LINE)
    print("  Goodbye!")
    print(LINE)
 
def print_history(expenses):
    if not expenses:
        print("  [!] No entries yet.\n")
        return
    print()
    print(f"  {'#':<5} {'Amount':>12}   {'Running Total':>14}")
    print("  " + "-" * 35)
    running = 0
    for i, e in enumerate(expenses, 1):
        running += e
        print(f"  {i:<5} Rs. {e:>8,.0f}   Rs. {running:>10,.0f}")
    print("  " + "-" * 35)
    print(f"  {'Total':<5} {'':>12}   Rs. {running:>10,.2f}")
    print()
 
# ── MAIN PROGRAM ─────────────────────────────────────────────
 
print_header()
 
expenses = []   # stores every valid entry
 
while True:
 
    user_input = input("  Enter expense (or command): ").strip()
 
    # ── Empty input ───────────────────────────────────────────
    if user_input == "":
        print("  [!] Nothing entered. Try a number or 'quit'.\n")
        continue
 
    cmd = user_input.lower()
 
    # ── Kill switch ───────────────────────────────────────────
    if cmd == "quit":
        print("  [Kill switch activated — session closing...]")
        break
 
    # ── History ───────────────────────────────────────────────
    if cmd == "history":
        print_history(expenses)
        continue
 
    # ── Undo last entry ───────────────────────────────────────
    if cmd == "undo":
        if not expenses:
            print("  [!] Nothing to undo.\n")
        else:
            removed = expenses.pop()
            total   = sum(expenses)
            print(f"  Removed: Rs. {removed:,}  |  New total: Rs. {total:,}\n")
        continue
 
    # ── Reset all ─────────────────────────────────────────────
    if cmd == "reset":
        expenses.clear()
        print("  All entries cleared. Starting fresh.\n")
        continue
 
    # ── Gatekeeper: str → float (Defensive Coding) ───────────
    try:
        expense = float(user_input)
 
        if expense <= 0:
            print("  [!] Amount must be greater than zero.\n")
            continue
 
        expenses.append(expense)
        total = sum(expenses)
        count = len(expenses)
 
        print(f"  Added : Rs. {expense:>8,.0f}  |  "
              f"Total: Rs. {total:,.0f}  |  "
              f"Entries: {count}\n")
 
    except ValueError:
        print(f"  [!] '{user_input}' is not a valid number. Numbers only!\n")
 
# ── Final Report ──────────────────────────────────────────────
print_report(expenses)
```
